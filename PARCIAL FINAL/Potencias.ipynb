```python
import numpy as np
import matplotlib.pyplot as plt
import math

xn=np.array([1, 2, 3, 4, 5, 6, 7, 8])

yn=np.array([3.8, 4, 4.2, 4.6, 4.9, 5.3, 5.7, 6.1])

lny=np.log10(yn)
lnx=np.log10(xn)

print(lny)
print(lnx)

n=len(xn)

Sum_x=sum(lnx)
Sum_y=sum(lny)
Sum_xx=sum(lnx**2)
Sum_xy=sum(lnx*lny)

print(Sum_x, Sum_y, Sum_xx, Sum_xy)
```

```python
from math import e

a_1=(n*Sum_xy-Sum_x*Sum_y)/(n*Sum_xx-Sum_x**2)
a_0= (Sum_y/n)-(a_1*Sum_x/n)

print(a_1)
print(a_0)

α=10**(a_0)
β=a_1

print(α)
print(a_1)

x=np.linspace(0, 9)
y=α*(x**a_1)

plt.figure(1)

plt.scatter(xn, yn, color='r')
plt.grid(linestyle='dotted')
plt.plot(x, y, color='b')

plt.title(r'Regresión por Mínimos', fontsize=10)
plt.xlabel(r'x', fontsize=10)
plt.ylabel(r'y', fontsize=10)

plt.show()
```
