```python
import numpy as np
import matplotlib.pyplot as plt

xn=np.array([1, 2, 3, 4, 5])

yn=np.array([5, 6.6, 8.5, 10.8, 14.2])

lny=np.log(yn)

print(lny)

n=len(xn)

Sum_x=sum(xn)
Sum_y=sum(lny)
Sum_xx=sum(xn**2)
Sum_xy=sum(xn*lny)

print(Sum_x, Sum_y, Sum_xx, Sum_xy)
```

```python
from math import e

a_1=(n*Sum_xy-Sum_x*Sum_y)/(n*Sum_xx-Sum_x**2)
a_0= (Sum_y/n)-(a_1*Sum_x/n)

print(a_1)
print(a_0)

α=e**(a_0)
print(α)

x=np.linspace(0, 5)
y=α*np.e**(a_1*x)

plt.figure(1)

plt.scatter(xn, yn, color='r')
plt.grid(linestyle='dotted')
plt.plot(x, y, color='b')

plt.title(r'Regresión por Mínimos', fontsize=10)
plt.xlabel(r'x', fontsize=10)
plt.ylabel(r'y', fontsize=10)

plt.show()
```
