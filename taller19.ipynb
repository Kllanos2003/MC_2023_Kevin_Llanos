```python
import numpy as np
import matplotlib.pyplot as plt

xn=np.array([1, 2, 3, 4, 5, 6, 7])

yn=np.array([0.2, 0.5, 1.8, 3.4, 5.7, 9.0, 13.8])
#yn=np.array([0.5, 2.5, 2, 4, 3.5, 6, 5.5])

n=len(xn)

Sum_x=sum(xn)
Sum_y=sum(yn)
Sum_xx=sum(xn**2)
Sum_xy=sum(xn*yn)

print(Sum_x, Sum_y, Sum_xx, Sum_xy)
```

```python
a_0=(Sum_xx*Sum_y-Sum_xy*Sum_x)/(n*Sum_xx-Sum_x**2)
a_1=(n*Sum_xy-Sum_x*Sum_y)/(n*Sum_xx-Sum_x**2)
print("a_0 y a_1 : ", a_0, a_1)


# Desviación estándar
std_dev = np.std(yn)

# Error estándar de la estimación
y_est = a_0 + a_1 * xn
residuos = yn - y_est
SSE = np.sum(residuos**2)
MSE = SSE / (n - 2)
std_error = np.sqrt(MSE)

# Coeficiente de correlación
correlation_coef = np.corrcoef(xn, yn)[0, 1]

x=np.linspace(0, 9)
y=a_0+a_1*x

plt.figure(1)

plt.scatter(xn, yn, color='r')
plt.grid(linestyle='dotted')
plt.plot(x, y, color='b')

plt.title(r'Regresión por Mínimos', fontsize=10)
plt.xlabel(r'x', fontsize=10)
plt.ylabel(r'y', fontsize=10)

plt.text(0.5, 10.5, f'Desviación estándar: {std_dev:.2f}')
plt.text(0.5, 9.5, f'Error estándar de la estimación: {std_error:.2f}')
plt.text(0.5, 8.5, f'Coeficiente de correlación: {correlation_coef:.2f}')

plt.show()
```
